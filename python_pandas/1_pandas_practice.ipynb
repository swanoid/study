## 1. 데이터 생성

### 1) 시리즈 생성: pd.Series() 함수 사용
- 리스트, 딕셔너리, 넘파이 배열 등을 입력 데이터로 사용 가능
- 1차원 배열 형태, 숫자 또는 레이블 인덱스(index) 지정 가능
- 하나의 열(column) 또는 하나의 행(row)처럼 동작

```python
import pandas as pd

# 리스트 사용 시리즈 생성
data = [10, 20, 30, 40]
series1 = pd.Series(data)
print(series1)
```

```python
# 인덱스 지정 시리즈 생성
index_labels = ['a', 'b', 'c', 'd']
series2 = pd.Series(data, index=index_labels)
print(series2)
```

```python
# 딕셔너리 사용 시리즈 생성 = 리스트 + 인덱스 지정
data_dict = {'A': 100, 'B': 200, 'C': 300}
series3 = pd.Series(data_dict)
print(series3)
```

### 2) 데이터프레임 생성: pd.DataFrame() 사용
- 딕셔너리, 리스트, 넘파이 배열, csv 파일 등 다양한 데이터 소스를 기반으로 생성
- 2차원 행렬 형태, 여러 개의 시리즈가 모여 하나의 데이터프레임 구성
- 각 행에는 인덱스(index), 각 열에는 컬럼(column) 이름 부여

```python
# 딕셔너리 사용 데이터프레임 생성
data_dict = {
    '이름': ['김철수', '이영희', '박민수'],
    '나이': [25, 30, 22],
    '성별': ['남', '여', '남']
}
df1 = pd.DataFrame(data_dict)
print(df1)
```

```python
# 리스트 사용 데이터프레임 생성
data_list = [
    ['김철수', 25, '남'],
    ['이영희', 30, '여'],
    ['박민수', 22, '남']
]
columns = ['이름', '나이', '성별']
df2 = pd.DataFrame(data_list, columns=columns)
print(df2)
```

## 2. 데이터 선택

### 1) 기본 인덱싱(시리즈, 데이터프레임)

```python
# 시리즈 기본 인덱스: 리스트와 유사한 형태 동작, 딕셔너리처럼 키-값 쌍 사용 가능
# 시리즈 생성
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print(s)
```

```python
# 기본 인덱스를 사용한 선택
print(s[0])
```

```python
# 레이블을 사용한 선택
print(s['a'])
```

```python
# 여러 개의 값 선택
print(s[['a', 'c']])
```

```python
# 데이터프레임 기본 인덱스: 행 또는 열 이름, 인덱스를 사용하여 특정 행이나 열 선택 가능
# 데이터프레임 생성
df = pd.DataFrame({
    '이름': ['김철수', '이영희', '박민수'],
    '나이': [25, 30, 22],
    '성별': ['남', '여', '남'],
    '점수': [90, 85, 95]
}, index = ['a', 'b', 'c'])
print(df)
```

```python
# 특정 열 선택
print(df['이름'])
```

```python
# 여러 열 선택
print(df[['이름', '나이']])
```

```python
# 첫 번째, 두 번째 행 선택
print(df[:2])
```

### 2) loc[]: 레이블 기반 선택
- 데이터프레임에서 행과 열의 레이블(label)을 기준으로 데이터 선택

```python
# 'a' 행 선택
print(df.loc['a'])
```

```python
# '이름' 열 선택
print(df.loc[:, '이름'])
```

```python
# 'a' 행과 '이름' 열 선택
print(df.loc['a', '이름'])
```

```python
# 여러 개의 행과 열 선택
print(df.loc[['a', 'b'], ['이름', '나이']])
```

### 3) iloc[]: 위치(인덱스) 기반 선택
- 데이터프레임에서 행과 열의 위치(position, 숫자 인덱스)를 기준으로 데이터 선택

```python
# 첫 번째 행 선택
print(df.iloc[0])
```

```python
# 두 번째 열 선택
print(df.iloc[:, 1])
```

```python
# 첫 번째 행과 두 번째 열 선택
print(df.iloc[0, 1])
```

```python
# 첫 번째, 두 번째 행과 첫 번째, 세 번째 열 선택
print(df.iloc[[0, 1], [0, 2]])
```

### 4) 조건 기반 선택

```python
# 특정 조건을 만족하는 행 선택
# 나이가 25보다 큰 사람만 선택
print(df[df['나이'] > 25])
```

```python
# 여러 조건을 사용한 필터링
# 나이가 20 초과이고 남자인 사람
print(df[(df['나이'] > 20) & (df['성별'] == '남')])
```

```python
# 특정 열의 값이 특정 리스트에 포함되는지 확인
# '김철수' 또는 '박민수'인 사람
print(df[df['이름'].isin(['김철수', '박민수'])])
```

```python
# 특정 조건을 만족하는 데이터만 선택
# 나이가 25 이상인 사람만 선택
filtered_df1 = df[df['나이'] >= 25]
print(filtered_df1)
```

```python
# 여러 조건을 적용한 필터링
# 나이가 25 이상이고 점수가 90 이상인 경우
filtered_df2 = df[(df['나이'] >= 25) & (df['점수'] >= 90)]
print(filtered_df2)
```

### 3) 데이터 변환: 데이터 가공하여 새로운 형식으로 변환

```python
# 새로운 열 추가
# '합격 여부' 열 추가: 점수가 90 이상이면 '합격', 아니면 '불합격'
df['합격 여부'] = df['점수'].apply(lambda x: '합격' if x >= 90 else '불합격')
print(df)
```

### 4) 결측치 처리
- 결측값 위치 확인: .isnull()
- 각 열 결측값 개수 확인: .isnull().sum()
- 결측값 포함 행 제거: .dropna()
- 결측값 채우기: .fillna()

```python
# 결측치 데이터 저장
df_missing = pd.DataFrame({
    '이름': ['김철수', '이영희', '박민수'],
    '나이': [25, None, 22],
    '점수': [90, 85, None]
})
print(df_missing)
```

```python
# 결측값이 있는 위치 확인
print(df_missing.isnull())
```

```python
# 각 열의 결측값 개수 확인
print(df_missing.isnull().sum())
```

```python
# 결측값이 포함된 행 제거
df_dropped = df_missing.dropna()
print(df_dropped)
```

```python
# 결측치 채우기: 평균 나이로 채우고 점수는 0으로 채움
df_filled = df_missing.fillna({'나이': df_missing['나이'].mean(), '점수': 0})
print(df_filled)
```

### 5) 중복 데이터 처리
- 중복 데이터 확인: .duplicated()
- 중복 데이터 제거: .drop_duplicates()

```python
# 중복 데이터 저장
df_duplicates = pd.DataFrame({
    '이름': ['김철수', '이영희', '김철수'],
    '나이': [25, 30, 25],
    '점수': [90, 85, 90]
})
print(df_duplicates)
```

```python
# 중복 데이터 확인
print(df_duplicates.duplicated())
```

```python
# 중복 데이터 제거
df_no_duplicates = df_duplicates.drop_duplicates()
print(df_no_duplicates)
```

### 4) 데이터 요약 및 집계
- .sum(), .mean(), .max(), .min()
- .groupby()

```python
data = {
    '카테고리': ['음식', '음식', '전자', '전자', '의류', '의류', '음식', '전자', '의류', '음식'],
    '매출': [120000, 150000, 300000, 250000, 180000, 200000, 170000, 320000, 220000, 140000]
}

# 데이터프레임 생성
data_df = pd.DataFrame(data)
print(data_df)
```

```python
# 데이터 요약
print(data_df['매출'].sum()) # 총 매출액
print(data_df['매출'].mean()) # 평균 매출액
print(data_df['매출'].max()) # 최대 매출액
print(data_df['매출'].min()) # 최소 매출액
```

```python
# 그룹별 요약(groupby)
grouped = data_df.groupby('카테고리')['매출'].sum()
print(grouped)
```

## 5. 데이터 편집

### 1) 데이터 수정: 데이터 가공의 데이터 필터링과 유사
- 특정 값 수정: .at[] (레이블), .iat[] (인덱스)
- 특정 열 값 일괄 수정
- 특정 조건 만족 행 값만 변경: .loc[] (레이블)
- 여러 개 열 수정: .loc[] (레이블)

```python
# 특정 값 수정
# 'a' 행의 '점수' 값을 88로 변경
df.at['a', '점수'] = 88
print(df)
```

```python
# 기존 인덱스 기본 숫자로 변경
reset_df = df.reset_index(drop=True)
print(reset_df)
```

```python
# 특정 값 수정
# 0번 행의 '점수' 값을 90으로 변경
reset_df.at[0, '점수'] = 90
print(reset_df)
```

```python
# 특정 값 수정
# 1번 행의 1번째(나이) 값을 32로 변경
df.iat[1, 1] = 32
print(df)
```

```python
# 특정 열 값 일괄 수정
# 모든 점수 변경
df['점수'] = [85, 80, 90]
print(df)
```

```python
# 특정 조건을 만족하는 행의 값만 변경
# 이름이 '김철수'인 행의 점수를 95로 변경
df.loc[df['이름'] == '김철수', '점수'] = 95
print(df)
```

```python
# 여러 개의 열 수정
df.loc[df['이름'] == '이영희', ['나이', '점수']] = [35, 88]
print(df)
```

### 2) 데이터 추가: 데이터 가공의 데이터 변환, 데이터 필터링과 유사
- 조건에 따른 새로운 열 추가: .apply()
- 새로운 열 직접 추가
- 새로운 행 직접 추가: .loc[]

```python
# 조건에 따른 새로운 열 추가(데이터 변환 참조)
```

```python
# 새로운 열 직접 추가
df['학교'] = ['서울대', '연세대', '고려대']
print(df)
```

```python
# 새로운 행 직접 추가
# .loc[len(df)] : 딱 그 번호를 의미함! 
df.loc[len(df)] = ['최민호', 28, '남', 87, '합격', '한양대']
print(df)
```

### 3) 데이터 삭제: 데이터 가공의 데이터 필터링과 유사
- 특정 열 삭제: .drop()
- 특정 행 삭제: .drop()
- 조건을 만족하는 행 삭제

```python
# 특정 열 삭제
# '학교' 열 삭제
df.drop(columns=['학교'], inplace=True)
print(df)
```

```python
# 특정 행 삭제
# 3번 인덱스 행 삭제: 딱 3번 인덱스 삭제!
df.drop(index=3, inplace=True)
print(df)
```

```python
# 특정 행 삭제
# 'b' 인덱스 행 삭제
df.drop(index='b', inplace=True)
print(df)
```

```python
# 조건을 만족하는 행 삭제
# 점수가 95 미만인 행 삭제
df = df[df['점수'] >= 95].copy()
print(df)
```

### 4) 데이터 변환
- 데이터 형식 변환: .astype() * 데이터 타입 확인: .dtypes
- 문자열 내 값 다른 값으로 변경: .str.replace()

```python
# 데이터 형식 변환
# 나이를 문자열로, 점수를 실수형으로 변환
df['나이'] = df['나이'].astype(str)
df['점수'] = df['점수'].astype(float)
print(df.dtypes)
```

```python
# 문자열 변환
# '김'을 'K'로 변경
df['이름'] = df['이름'].str.replace('김', 'K')
print(df)
```
