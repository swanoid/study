```python
# Shapiro_Wilk 정규성(정규분포) 검정
shapiro_test = stats.shapiro(df['median_house_value'].sample(500))
print(f"Shapiro_Wilk Test Statistic: {shapiro_test.statistic}, p_value: {shapiro_test.pvalue}")
```

```python
# t-검정
t_test_result = stats.ttest_1samp(df['median_house_value'].sample(500), 200000)
print(f"t-Statistic: {t_test_result.statistic}, p-value: {t_test_result.pvalue}")
```

```python
from scipy.optimize import curve_fit
import numpy as np

# 선형관계를 모델링하는 함수 정의
def linear_model(x, a, b):
  return a * x + b

# 예를 들어, 'median_income'과 'median_house_value' 관계를 살펴본다
x_data = df['median_income']
y_data = df['median_house_value']

# curve_fit 함수를 사용해 모델 파라미터 최적화
params, covariance = curve_fit(linear_model, x_data, y_data)

print(f"Optimized parameters: {params}")
```

```python
from scipy.signal import convolve
import matplotlib.pyplot as plt

median_house_value = df['median_house_value']

# 이동 평균을 계산하기 위한 윈도 생성
window_size = 5
window = np.ones(window_size) / window_size

# 이동 평균 필터 적용
smoothed_signal = convolve(median_house_value, window, mode = 'valid')

# 원본 신호와 스무딩 신호 비교
plt.figure(figsize = (10, 6))
plt.plot(median_house_value.index, median_house_value, label = 'Original')
plt.plot(np.arange(window_size -1, len(median_house_value)), smoothed_signal, label = 'Smoothed', color = 'red')
plt.legend()
plt.show()
```

# Statsmodels

```python
# !pip install statsmodels
import pandas as pd
import statsmodels.api as sm

# 데이터 로드
housing_data = pd.read_csv('sample_data/california_housing_test.csv')

# 독립변수와 종속변수 선택
X = housing_data['median_income'] # 독립변수
y = housing_data['median_house_value'] # 종속변수

# 상수항 추가
X = sm.add_constant(X)
```

```python
# 선형회귀모델 피팅
model = sm.OLS(y, X).fit()

# 결과 요약 출력
print(model.summary())
```

```python
# 새로운 데이터 준비
# print(df['median_income'].describe())
# 예를 들어, median_income이 2.5, 4.5, 6.5인 세 단지에 대한 중위 주택 가격을 예측하고자 한다.
new_data = pd.DataFrame({'const': 1, 'median_income': [2.5, 4.5, 6.5]})

# 예측 수행
predictions = model.predict(new_data)

print(predictions)
```

# 데이터과학 프로젝트
- 목표 설정: 문제 정의, 프로젝트 방향성 결정 ex) 서울시 보행 친화적 교통환경 조성 기여(목표) -> 서울 시내 도로 주말 동안 보행자 전용도로로 전환(정책) -> 주말에 보행자가 더 많이 통행하는지, 주말-평일 보행자 통행량 비교(타당성 검증)
- 데이터 수집: 공개 데이터셋, 조직의 내부 데이터, 온라인 설문조사, 웹 스크래핑
- 데이터 준비(ETL): 데이터 정제(결측치 처리, 이상치 제거), 데이터 변환(형식 변경, 정규화, 범주형 데이터 인코딩), 특성 선택 및 생성(방문자수, 주말여부, 요일)
- 데이터 탐색(EDA): 데이터 다양한 각도에서 조사, 데이터 패턴, 이상 현상, 특성 간의 관계 파악 ex) 시각화(상자-수염 그림, 그래프(요일, 주중/주말 보행자수 차이)
- 데이터 모델링: 통계적 모델 및 머신러닝 알고리즘 적용, 모델의 성능 평가 및 비교, 최적의 모델 선택 ex) 정규성 검정(Shapiro-Wilk test): 정규분포 평가(정규성 검정 유의미하면(p값 작으면), 데이터 정규분포 따르지 않는다고 간주 -> 적절한 통계 검정(z-test or t-test) 수행
- 활용 및 배포: 최종 모델 실제 환경에 적용, 예측, 추천, 자동화 등 작업 수행, 모니터링, 업데이트

```python
# 주중/주말 보행자수 추이
sns.barplot(data = walk_df, x = '주말여부', y = '방문자수')
plt.xlabel('weekend')
plt.ylabel('traffic')
plt.show()
```

```python
# 주중/주말 평균 방문자수
weekend_df = walk_df[walk_df['주말여부']]
weekday_df = walk_df[~walk_df['주말여부']]

print('주말 일별 평균 방문자수', weekend_df['방문자수'].mean())
print('주중 일별 평균 방문자수', weekday_df['방문자수'].mean())
```

```python
# 데이터 모델링
from scipy.stats import shapiro

# 주중 데이터의 방문자수에 대한 Shapiro-Wilk 검정
weekday_visitors_normality = shapiro(weekday_df['방문자수'])

# 주말 데이터의 방문자수에 대한 Shapiro-Wilk 검정
weekend_visitors_normality = shapiro(weekend_df['방문자수'])

# 검정 결과 출력
print('주중 방문자수 정규성 검정 결과:', weekday_visitors_normality)
print('주말 방문자수 정규성 검정 결과:', weekend_visitors_normality)
```

- 정규성 띠지 않음, 상자-수염 그림에서 분포가 치우쳐 있어 예상할 수 있었음

```python
# Welch's t-test 수행: 두 집단의 분포 비교
# 귀무가설: 주중/주말의 평균 방문자수는 차이가 없다.
# 대립가설: 주중/주말의 평균 방문자수는 차이가 있다.
from scipy.stats import ttest_ind

# t-test 수행
t_stat, p_value = ttest_ind(weekday_df['방문자수'], weekend_df['방문자수'], equal_var = False)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

- t값: 주중-주말 데이터 간 평균 방문자수 차이가 0이라는 가설에 대한 표본 데이터의 호환성 나타냄 -> 주중이 주말 데이터 평균 방문자수보다 통계적으로 유의하게 낮다는 것 나타냄
- p값: 귀무가설이 참일 떄, 현재 표본 데이터 또는 더 극단적 결과를 얻을 확률 -> p값 매우 낮으므로(0.05, 5% 기준) 관측 데이터가 귀무가설과 일치하지 않음, 주중-주말 방문자수 사이 통계적 유의미한 차이 있음
- 결론: 주말에는 공공장소 등에서의 보행자수가 유의미하게 증가한다.
- 위협1: 1주일간 데이터 사용, 특별 이벤트 기간일 수 있음, 해결1: 더 오랜 기간 데이터 사용
- 위협2: 수집 데이터 거주지역, 업무지역인지에 따라서도 영향, 해결2: 지역, 행정동 외부 데이터와 결합함으로써 판단 가능
- 위협3: 상관관계 보여줄 수 있지만 인과관계 단정 짓기 어려움, 해결3: 인과분석(casual analysis) 이용하여 주중/주말 여부가 방문자수에 영향을 미치는 주된 요인인지 판단
